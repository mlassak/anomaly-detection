```python
from data_utils.csv_utils import read_timeseries_csv
```

```python
pod_1_df = read_timeseries_csv("./examples/sample_data/pod_level_data/cpu_usage_pod_1.csv")
pod_2_df = read_timeseries_csv("./examples/sample_data/pod_level_data/cpu_usage_pod_2_first.csv")
pod_2_replacement_df = read_timeseries_csv("./examples/sample_data/pod_level_data/cpu_usage_pod_2_replacement.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(pod_1_df["value"], label="Pod 1")


plt.xlabel("Timestamp")
plt.ylabel("CPU usage %")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(pod_2_df["value"], label="Pod 2", color="orange")
plt.plot(pod_2_replacement_df["value"], label="Pod 2 replacement", color="green")


plt.xlabel("Timestamp")
plt.ylabel("CPU usage %")

plt.legend()
plt.tight_layout()
plt.show()
```
